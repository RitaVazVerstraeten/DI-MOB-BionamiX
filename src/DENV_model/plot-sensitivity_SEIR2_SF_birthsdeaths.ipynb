## Plotting the results of the sensitivity analysis for the temporal SEIR2 model with seasonal forcing and births and deaths

The plot in this file is based on Tijs Alleman's plot on GitHub: 
https://github.com/UGentBiomath/COVID19-Model/blob/master/notebooks/calibration/plot_sensitivity-COVID19_SEIQRD_hybrid_vacc.py

```python
# load the dependencies
import pandas as pd
import matplotlib.pyplot as plt
import os

from sensitivity import plot_sobol_indices_bar, plot_sobol_indices_circular
```

```python
# Load data
results_folder = '/home/rita/PyProjects/DI-MOB-BionamiX/results/DENV_model/Temporal_SEIR2_SF_birthdeath_test/sobol_sensitivity/sobol_temp_SEIR2_2024-11-26.xlsx'
S1ST = pd.read_excel(results_folder, sheet_name='S1ST', index_col=[0])
S2 = pd.read_excel(results_folder, sheet_name='S2', index_col=[0])

# S1ST = S1ST.reset_index()
```

```python
# Hardcoded labels
labels = ['$\\alpha$','$b$', '$d$', '$\\sigma$', '$\\gamma$', '$\\omega$', '$\\psi$', '$\\beta_0$', '$\\beta_1$', '$\\phi$']
print(labels)
```

```python
# Bar plot
fig,ax=plt.subplots(figsize=(12,4))
ax = plot_sobol_indices_bar(S1ST, ax, labels)
plt.tight_layout()
plt.show()
plt.close()

base_name = os.path.splitext(os.path.basename(results_folder))[0]
# Create the new file path by appending '_barplot' to the base name
new_file_path = os.path.join(os.path.dirname(results_folder), base_name + '_barplot.png')
 # Save the figure
ax.figure.savefig(new_file_path, bbox_inches='tight')
plt.close()

```

```python
# Circular plot
fig=plot_sobol_indices_circular(S1ST, S2, labels)
plt.tight_layout()
plt.show()
plt.close()

new_file_path = os.path.join(os.path.dirname(results_folder), base_name + '_circularplot.png')
 # Save the figure
fig.savefig(new_file_path, bbox_inches='tight')
```
