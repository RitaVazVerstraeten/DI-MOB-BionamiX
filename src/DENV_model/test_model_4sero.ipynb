# Test model_4 serotypes - simple no pySODM
Simple notebook to test the calculate_rate() and apply_transitionings() functions separately before using them within the pySODM framework. 
1. calculate_rates() based on state_values and params
2. calculate the transitionings as rates*states_values (so deterministic at the moment)
3. apply_transitionings() using transitionings and state_values

Seems to be working!

```python
import numpy as np
import importlib
import model_4sero
importlib.reload(model_4sero)
from model_4sero import JumpProcess_SEIR4_beta_by_Temp_sf_BirthDeath_reported
```

```python
# Get state order and mapping
states_list = JumpProcess_SEIR4_beta_by_Temp_sf_BirthDeath_reported.states_list
state_name_to_idx = JumpProcess_SEIR4_beta_by_Temp_sf_BirthDeath_reported.state_name_to_idx
```

```python
# 1. Create test state values: S=9960, I1=20, I2=20, all others zero (total 10000)
state_values = []
for state in states_list:
    if state == "S":
        state_values.append(9960)
    elif state == "I1" or state == "I2":
        state_values.append(20)
    else:
        state_values.append(0)

print("State values:", states_list, "\n", state_values)
print(len(state_values), "states in total.")
```

```python
# 2. Set parameter values
params = dict(
    b=2.77e-05, d=2.45e-05, beta_0=0.5, alpha=365/2, sigma=7, gamma=7, psi=0.5, rho=0.1, sf=1
)
```

```python
# 3. Test compute_rates
rates = JumpProcess_SEIR4_beta_by_Temp_sf_BirthDeath_reported.compute_rates(
    0, *state_values, **params)

print("\n--- Rates for S, I1 and I2---")
print("S:", rates["S"])
print("I1:", rates["I1"])
print("I2:", rates["I2"])
```

```python
# # 2. Create a dummy transitionings dict: all zeros except a few test transitions
# transitionings = {state: [0]*6 for state in states_list}  # 6 is max needed for S
# transitionings["S"][0] = 2   # S -> E1
# transitionings["S"][1] = 3   # S -> E2
# transitionings["S"][5] = 1   # births into S
# # All other transitions remain zero


# 4. Calculate transitionings based on rates and state values
transitionings = {}
for state in states_list:
    idx = state_name_to_idx[state]
    if state in rates:
        print(f"Processing state: {state}")
        print("State value:", state_values[idx])
        print("Rates for state:", rates[state])
        transitionings[state] = [rate * state_values[idx] for rate in rates[state]]
        print("Transitionings:", transitionings[state])
    else:
        # For derivative states, set transitions to [0] or an empty list as appropriate
        print(f"Processing derivative state: {state} (index {idx})")
        transitionings[state] = [0]
```

```python
print("\n--- Prior state values for S, I1, I2 ---")
print("S_new:", state_values[state_name_to_idx["S"]])
print("E1_new:", state_values[state_name_to_idx["E1"]])
print("E2_new:", state_values[state_name_to_idx["E2"]])
print("I1_new:", state_values[state_name_to_idx["I1"]])
print("I2_new:", state_values[state_name_to_idx["I2"]])
print("R1_new:", state_values[state_name_to_idx["R1"]])
print("R2_new:", state_values[state_name_to_idx["R2"]])

# 5. Test apply_transitionings
new_states = JumpProcess_SEIR4_beta_by_Temp_sf_BirthDeath_reported.apply_transitionings(
    0, 1, transitionings, *state_values, **params
)
print("\n--- New state values for S, I1, I2 ---")
print("S_new:", new_states[state_name_to_idx["S"]])
print("E1_new:", new_states[state_name_to_idx["E1"]])
print("E2_new:", new_states[state_name_to_idx["E2"]])

print("I1_new:", new_states[state_name_to_idx["I1"]])
print("I2_new:", new_states[state_name_to_idx["I2"]])
print("R1_new:", new_states[state_name_to_idx["R1"]])
print("R2_new:", new_states[state_name_to_idx["R2"]])
```

## Let's simulate all 4 scaling factors within the 4 serotype model and plot them

```python
# define function to intialize and simulate the model - with choice of scaling factor
def init_sim_model(scaling_factor_name, scaling_factor_df, time_unit = "D"):
    '''
    Function where the model get's initialized and simulated using the correct scaling factor values 
    Beta(t) is converted into beta_0 x sf through the time_dependent_beta function 
    
    Input 
    -----
    scaling_factor_model: string containing the name of the model. Can be model_SF, model_Lamb, model_Perk, model_Mord

    scaling_factor_SF: string containing the name under which the scaling factors are stored. Can be "seasonal_forcing", "mordecai_aeg", "lambrechts", or "perkins".

    Output
    ------
    Model simulation in xarray.Dataset
    '''
    # set the parameters
    params={'alpha':182.5, 'b':2.77e-05, 'd':2.45e-05, 'sigma':6, 'gamma':7, 'psi': 1.5, 'beta_0' : 0.25, 'sf' : scaling_factor_df, 'rho' : 0.40} 

    # if the time_unit is set to weeks then convert the parameters related to rates into weeks instead of days 
    if time_unit == "W":
        params_week = convert_params_to_weeks(params)
        params = params_week
    # set the initial conditions
    initial_infected = 80
    # Equal probabilities for 4 primary infectious compartments
    probabilities = [0.25, 0.25, 0.25, 0.25]

    # Stochastic division using multinomial
    [init_I1, init_I2, init_I3, init_I3] = np.random.multinomial(initial_infected, probabilities)
    # print("init_I1", init_I1, "\ninit_I2", init_I2, "\ninit_I12", init_I12, "\ninit_I21", init_I21)

    # Define initial condition
    initN = 100000
    initS = initN - (init_I1 + init_I2 + init_I3 + init_I3)

    initial_states = {
        'S': np.array([initS]),
        'I1': np.array([init_I1]),
        'I2': np.array([init_I2]),
        'I3': np.array([init_I3]),
        'I4': np.array([init_I3])
    }
    
    # print("\nInitial states", initial_states)

    # initialize the model 
    model = model_4sero(initial_states=initial_states, parameters=params, 
            time_dependent_parameters={"beta_0": time_dep_beta})

    # simulate the model + draw_fnc
    out = model.sim(time=[start_date, end_date],N=n, tau=tau, time_unit = time_unit)
    # out = model.sim(time = ["2012-06-01", "2013-05-01"], N =n , tau=tau, output_timestep=7)

    return out
```

```python
# function to plot the results per SF type
def plot_simulation_results(simulation_results, plot_title):
    """
    Plots I_rep over time for each model in a 2x2 grid, showing all draws

    Parameters:
    - simulation_results: dict, where keys are model names and values contain simulation outputs 
      with dimensions (date, draws).

    Returns:
    - fig: matplotlib figure object.
    """
    num_models = len(simulation_results)
    fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))  # 2x2 grid
    axes = axes.flatten()  # Flatten to iterate easily

    for idx, (model_name, results) in enumerate(simulation_results.items()):
        ax = axes[idx]
        dates = results["date"].values  # Convert xarray DataArray to NumPy array
        num_draws = results["I_rep"].shape[0]  # Number of draws

        # Plot individual trajectories for I_rep
        for i in range(num_draws):
            ax.plot(dates, results["I_rep"][i, :], color="red", alpha=0.1, label="Draws" if i == 0 else "")

        # Overlay mean trend for I_rep
        ax.plot(dates, np.mean(results["I_rep"], axis=0), color="blue", linewidth=2, label="Mean I_rep")

        # Formatting
        ax.set_title(f"Simulation Results for {model_name}")
        ax.set_xlabel("Date")
        ax.set_ylabel("Reported Infections (I_rep)")
        ax.tick_params(axis='x', rotation=45)
        ax.grid(True)

        # Add legend
        ax.legend(loc="upper right")  # Add legend in the top right corner
    plt.suptitle(f"{plot_title}")

    # Adjust layout for better spacing
    plt.tight_layout()
    return fig  # Return figure object
```

```python
# Loop over all the scaling factors and store the model simulation results
tau = 1.0
time_unit = "W"
n = 50 # repetitions for stochastic simulation

simulation_results_tau1_unitW = {}

for model_name, df  in scaling_factors.items():
    print("\n Simulating for", model_name)
    sim_result = init_sim_model(scaling_factor_name = model_name, scaling_factor_df = df, time_unit="W")
    # Store results
    simulation_results_tau1_unitW[model_name] = sim_result
    print(f"\nSimulation completed for {model_name} with {n} iterations")
```

```python
fig = plot_simulation_results(simulation_results_tau1_unitW, plot_title="Test 4-sero model 1 year period")
```
